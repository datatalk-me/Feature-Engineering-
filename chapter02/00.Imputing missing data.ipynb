```python
# importing required libraries
import random
import numpy as np
import pandas as pd
```

```python
# loading the data

data = pd.read_csv('data/crx.data',header=None)
data.head()
```

```python
# create a list with variable names
variablenames = ['A'+str(s) for s in range(1,17)]
variablenames
```

```python
# adding the varibale names for the dataframe
data.columns = variablenames
```

```python
data.head()
```

```python
# replacing the question marks with nan values
data = data.replace('?',np.nan)
data
```

```python
data.info()
```

```python
# recast the numerical variables as float data types
data['A2'] = data['A2'].astype('float')
data['A14'] = data['A14'].astype('float')
```

```python
data.info()
```

```python
# recode the target variable as binary
data['A16'] = data['A16'].map({'+':1,'-':0})
data.head()
```

To demonstrate the recipes in this chapter, we will introduce missing data at
random in four additional variables in this dataset.

```python
# adding missing values at random
random.seed(9001)
```

```python
values = set([random.randint(0,len(data)) for p in range(0,100)])
```

```python
for var in ['A3','A8','A9','A10']:
    data.loc[values,var] = np.nan
```

```python
data.to_csv('data/creditApprovalUCI.csv',index=False)
```
